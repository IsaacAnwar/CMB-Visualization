```python
!pip install matplotlib

import matplotlib.pyplot as plt

# Parameters for the Sun (a star)
sun_core_temperature = 15.6e6  # in Kelvin
sun_core_pressure = 2.48e10   # in Pascals
sun_energy_release = 3.8e26    # in watts (Sun's total energy output)

# Parameters for a nuclear bomb
bomb_temperature = 1e8        # in Kelvin
bomb_pressure = 6.89e9          # in Pascals
bomb_energy_release = 3.38e25    # in watts

# Convert bomb's energy release to watts for consistency
bomb_energy_release_watts = bomb_energy_release / 1e6  # 1 watt = 1 joule/second

# Plotting
labels = ['Sun', 'Nuclear Bomb']
temperatures = [sun_core_temperature, bomb_temperature]
pressures = [sun_core_pressure, bomb_pressure]
energy_releases = [sun_energy_release, bomb_energy_release_watts]  # Convert to watts for consistency

fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(15, 5))

# Temperature comparison
ax1.bar(labels, temperatures, color=['gold', 'red'])
ax1.set_title('Temperature Comparison')
ax1.set_ylabel('Temperature (Kelvin)')

# Pressure comparison
ax2.bar(labels, pressures, color=['gold', 'red'])
ax2.set_title('Pressure Comparison')
ax2.set_ylabel('Pressure (Pascals)')

# Energy release comparison
ax3.bar(labels, energy_releases, color=['gold', 'red'])
ax3.set_title('Energy Release Comparison')
ax3.set_ylabel('Energy Release (Watts)')

plt.tight_layout()
plt.show()


```
